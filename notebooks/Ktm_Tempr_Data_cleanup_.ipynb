```python
df.set_index('Time', inplace=True)
df.sort_index(inplace=True) # Always good practice after setting index
```

```python
df
```

```python
# Filter for entries at 03:00:00 UTC
df_daily_min = df[df.index.time == pd.to_datetime('03:00:00').time()].copy()
df_daily_min   
```

```python
# Only keep the 'Min_Temp_C' and 'Precipitation_mm' columns
df_daily_min = df_daily_min[['Precipitation_mm', 'Min_Temp_C']]
```

```python
# Filter for entries at 03:00:00 UTC
df_daily_max = df[df.index.time == pd.to_datetime('12:00:00').time()].copy()
```

```python
# Only keep the 'Max_Temp_C' and 'Precipitation_mm' columns
df_daily_max = df_daily_max['Max_Temp_C']
```

```python
df_daily_max.head()
```

```python
# Coerce errors, which will turn non-numeric (like empty strings) into NaN
df['Precipitation_mm'] = pd.to_numeric(df['Precipitation_mm'], errors='coerce')
df['Max_Temp_C'] = pd.to_numeric(df['Max_Temp_C'], errors='coerce')
df['Min_Temp_C'] = pd.to_numeric(df['Min_Temp_C'], errors='coerce')
```

```python
df
```

```python
# Reset index to just date component for merging
df_daily_min.index = df_daily_min.index.normalize() # This converts timestamp to just date
df_daily_max.index = df_daily_max.index.normalize()

# Join the two daily dataframes. An outer join ensures all dates are kept.
df_daily_combined = df_daily_min.join(df_daily_max, how='outer')
```

```python
df_daily_combined
```

```python
print("--- Consolidated Daily Data Info ---")
df_daily_combined.info()
print("\n--- Missing Values in Consolidated Daily Data ---")
print(df_daily_combined.isnull().sum())
print("\n--- Sample of Consolidated Daily Data ---")
print(df_daily_combined.head())
print("\n--- Tail of Consolidated Daily Data ---")
print(df_daily_combined.tail())
```

```python
first_valid_temp_date = df_daily_combined[df_daily_combined['Max_Temp_C'].notna() & df_daily_combined['Min_Temp_C'].notna()].index.min()
print(f"\nFirst date with complete temperature data: {first_valid_temp_date}")
```

```python
df_analysis = df_daily_combined.loc[first_valid_temp_date:].copy()
```

```python
df_analysis
```

```python
# For temperature, linear interpolation is usually good for small gaps
df_daily_combined['Max_Temp_C'].interpolate(method='linear', inplace=True)
df_daily_combined['Min_Temp_C'].interpolate(method='linear', inplace=True)

# For precipitation, an isolated NaN often means no rain was recorded, so fill with 0
df_daily_combined['Precipitation_mm'].fillna(0, inplace=True)
```

```python
df_daily_combined
```

```python
df_daily_combined.info()
```

```python
print("\n--- Final Consolidated & Cleaned Data Info ---")
df_daily_combined.info()
print("\n--- Missing Values after Final Cleaning ---")
print(df_daily_combined.isnull().sum())
```

```python
df_analysis = df_daily_combined.copy()
```

# Daily Max and Min Air Temperature in Kathmandu (1999-2024)
This time-series plot reveals Kathmandu's daily temperature dynamics over 25 years.
- Dominant Annual Seasonality: 
Displays a very strong and consistent yearly cycle for both maximum (blue line) and minimum (orange dashed line) temperatures, reflecting clear summer and winter periods.
- Consistent Daily Fluctuation: 
The distinct gap between the maximum and minimum temperature lines throughout the year illustrates the daily diurnal range, showing how temperature fluctuates within a single 24-hour period.
- Inter-annual Variability: 
While the overall seasonal pattern is stable, there's noticeable year-to-year variability in the amplitude of temperature swings and specific daily values, indicating natural weather variations.
- Long-term Trend (Visual): 
Visually, a prominent, consistent linear warming or cooling trend is not immediately evident across the entire period, suggesting a relatively stable overall climate pattern within this dataset's timeframe. However, fluctuations exist year to year.
## Observed Extremes:
- Winter Cold: 
Minimum temperatures frequently drop close to 0°C during the coldest winter months (typically Dec-Feb).
- Summer Heat: 
Maximum temperatures consistently reach into the low 30s °C during the warmest summer months (typically Jun-Aug).

*Occasional sharp spikes or drops are evident, representing daily weather events that might be candidates for anomaly detection.*


```python
# For Precipitation
plt.figure(figsize=(35, 15))

# Calculate a 7-day rolling sum to smooth out daily noise and highlight wet periods
df_analysis['Precipitation_7d_rolling_sum'] = df_analysis['Precipitation_mm'].rolling(window=7).sum()

# Plot the data
sns.lineplot(x=df_analysis.index, y=df_analysis['Precipitation_7d_rolling_sum'], color='blue')

# Enhance the x-axis with more detailed formatting
plt.title('7-Day Rolling Sum of Accumulated Precipitation in Kathmandu (1999-2024)')
plt.xlabel('Date', fontsize=14, fontweight='bold')
plt.ylabel('Precipitation (mm/7 days)')

# Set x-axis ticks to show years clearly
years = pd.date_range(start=df_analysis.index.min(), end=df_analysis.index.max(), freq='YS')
plt.xticks(years, [year.strftime('%Y') for year in years], rotation=45, ha='right', fontsize=12)

# Add minor ticks for quarters
plt.gca().xaxis.set_minor_locator(plt.matplotlib.dates.MonthLocator(bymonth=[1, 4, 7, 10]))

# Add vertical lines to separate years for better readability
for year in years:
    plt.axvline(x=year, color='gray', linestyle='--', alpha=0.5)

# Add grid for better readability
plt.grid(True, which='major', linestyle='-', linewidth=0.5)
plt.grid(True, which='minor', linestyle=':', linewidth=0.3)

# Improve margins to ensure dates are fully visible
plt.tight_layout()
plt.savefig('precipitation')
plt.show()
```

# 7-Day Rolling Sum of Accumulated Precipitation in Kathmandu (1999-2024)
- Dominant Monsoon Seasonality: Shows a very strong and consistent annual cycle characterized by distinct wet and dry periods. The monsoon season is clearly visible with significant precipitation, primarily occurring from approximately June to September.
- Pronounced Dry Season: Outside the monsoon, precipitation is consistently low or non-existent, indicating a prolonged dry season, particularly from October through May.
- Variable Monsoon Intensity: The height and duration of the blue peaks vary considerably year-to-year. Some years exhibit higher and more sustained rainfall (e.g., 2000, 2002, 2010, 2020, 2024 (partial)), while others appear to have comparatively drier monsoons or shorter intense periods (e.g., 2004, 2007, 2008, 2014, 2017).
- Extreme Events: The sharp, tall spikes within the monsoon periods represent significant, concentrated rainfall events. These visually indicate instances of heavy downpours or extended wet spells, which are crucial for flood risk assessment and water resource management.
- No Obvious Long-Term Trend: Across the 25-year span, there isn't an immediately apparent linear increase or decrease in overall monsoon intensity or duration, though inter-annual variability is high

# Average Monthly Max/Min Temperature in Kathmandu (1999-2024)
This plot clearly illustrates Kathmandu's distinct annual temperature cycle, reflecting its subtropical highland climate.
- Prominent Annual Cycle: 
Temperatures follow a strong, predictable seasonal pattern, reflecting the Northern Hemisphere's winter and summer.
- Coldest Period: 
January and February are the coldest months (Min Avg: ~3-5°C; Max Avg: ~18-21°C), aligning with the peak winter season.
- Warmest Period: 
Temperatures steadily rise from March, peaking in June-July (Max Avg: ~29-30°C; Min Avg: ~20°C), corresponding to the pre-monsoon and early monsoon warmth.
## Diurnal Temperature Range (DTR) - Seasonal Variation:
- Wider DTR (Dry Season - Oct-May): 
The larger gap between Max and Min temperatures, particularly prominent in spring (Mar-Apr) and post-monsoon (Oct-Nov), indicates clearer skies and lower humidity. This allows for greater radiative cooling at night (cooler lows) and more intense solar heating during the day (warmer highs).
- Narrower DTR (Monsoon Season - Jun-Sep): 
The compressed temperature range during these months is a hallmark of monsoon influence. Increased cloud cover and high humidity dampen both daytime heating and nighttime cooling, resulting in consistently warmer nights and relatively subdued daytime highs.
- Climate Alignment: 
These observations are characteristic of a monsoon climate, where atmospheric moisture and cloud cover significantly regulate diurnal temperature swings.

# Total Monthly Precipitation in Kathmandu (1999-2024)
This bar chart clearly delineates the average monthly precipitation, further highlighting Kathmandu's strong monsoon climate.
- Dominant Monsoon Rainfall: The vast majority of precipitation is concentrated during the summer monsoon months, primarily from June to September.
- Peak Monsoon: July consistently stands out as the wettest month, accumulating the highest average rainfall, followed closely by August and then September. This confirms the core period of the South Asian monsoon.
- Distinct Dry Season: A very pronounced dry season is evident from October through May. Precipitation during these months is significantly low, with January, February, November, and December receiving negligible amounts.
- Pre-Monsoon Build-up: There's a gradual increase in rainfall in April and May, indicating the typical pre-monsoon warming and atmospheric destabilization before the full onset of the monsoon.
- Post-Monsoon Transition: Rainfall significantly drops off in October, marking the swift transition from the wet to the dry season, with minimal precipitation returning in November and December.

```python
# Create a figure with specified size (width=15, height=5 inches)
plt.figure(figsize=(15, 5))

# Create first subplot for maximum temperature distribution
plt.subplot(1, 3, 1)
sns.histplot(df_analysis['Max_Temp_C'], kde=True, bins=20)  # Plot histogram with kernel density estimate using 20 bins
plt.title('Distribution of Max Temperature')
plt.xlabel('Temperature (°C)')

# Create second subplot for minimum temperature distribution
plt.subplot(1, 3, 2)
sns.histplot(df_analysis['Min_Temp_C'], kde=True, bins=20)  # Plot histogram with kernel density estimate using 20 bins
plt.title('Distribution of Min Temperature')
plt.xlabel('Temperature (°C)')

# Create third subplot for precipitation distribution
plt.subplot(1, 3, 3)
# For precipitation, plotting only non-zero values or using a log transform is often best due to heavy skewness
sns.histplot(df_analysis[df_analysis['Precipitation_mm'] > 0]['Precipitation_mm'], kde=True, bins=20)  # Filter out zero values
plt.title('Distribution of Non-Zero Precipitation')
plt.xlabel('Precipitation (mm)')
plt.tight_layout()  # Adjust subplot parameters to give specified padding
plt.savefig('distribution.png')
plt.show()  
```

# Distribution of Max Temperature, Min Temperature, and Non-Zero Precipitation
These histograms, with their Kernel Density Estimates (KDEs), reveal the frequency and spread of temperature and precipitation values in Kathmandu.
##### Distribution of Max Temperature:
- The distribution is largely unimodal (single peak) and somewhat bell-shaped, centered around 28-30°C.
- This indicates that maximum temperatures in Kathmandu most frequently fall within a comfortable warm range, representing the prevalent conditions during most of the year outside of peak winter.
##### Distribution of Min Temperature:
- This plot reveals a bimodal (two-peak) distribution, which is highly insightful.
- One distinct peak occurs around 3-5°C, representing the cold minimum temperatures experienced during the winter months.
- A broader, less defined peak or plateau is visible around 18-20°C, reflecting the warmer minimum temperatures during the monsoon and post-monsoon seasons.
- This bimodality strongly indicates two distinct thermal regimes for nighttime temperatures in Kathmandu, primarily driven by seasonal atmospheric conditions (clear winter nights allowing strong radiative cooling vs. humid, cloudy monsoon nights trapping heat).
##### Distribution of Non-Zero Precipitation:
- The distribution is highly right-skewed (positively skewed), with a large concentration of data points near 0 mm.
- This confirms that while precipitation events are frequent, the vast majority are small in magnitude (e.g., 0-25 mm).
- The long tail extending to higher values indicates that very heavy rainfall events (e.g., >100 mm) occur but are comparatively rare, yet crucial for flood assessment.
- This extreme skewness implies that standard statistical models might benefit from data transformations (like a log transform) when working with precipitation, to better approximate a normal distribution.


```python
plt.figure(figsize=(8, 6))
sns.heatmap(df_analysis[['Max_Temp_C', 'Min_Temp_C', 'Precipitation_mm']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Key Variables')
plt.savefig('Heatmap.png')
plt.show()
```

# Correlation Matrix of Key Variables
This heatmap displays the pairwise linear correlation coefficients between Maximum Temperature, Minimum Temperature, and Daily Accumulated Precipitation.
- Strong Correlation between Max and Min Temperature (0.77): As anticipated, maximum and minimum daily temperatures are highly and positively correlated. This indicates that warmer days are generally followed by warmer nights, and colder days by colder nights, reflecting the overall seasonal temperature shifts.
- Weak Positive Correlation: Max Temperature vs. Precipitation (0.12): There is a very weak positive linear relationship between maximum temperature and precipitation over the entire year. This might seem counter-intuitive at first given that monsoons occur in the warmer months. However, this is an average linear correlation across all seasons. The pre-monsoon months are hot but dry, and the core monsoon months, while warm, often see reduced peak daytime temperatures due to cloud cover.
- Moderate Positive Correlation: Min Temperature vs. Precipitation (0.33): Minimum temperatures show a slightly stronger, but still moderate, positive linear correlation with precipitation. This could be influenced by the fact that nights during the monsoon season (when precipitation is high) are generally warmer due to increased humidity and cloud cover reducing radiative cooling.
- Implications for Modeling: While these linear correlations are informative, the relatively low values with precipitation suggest that a simple linear model might not fully capture the complex, non-linear, and seasonally varying relationship between temperature and rainfall in Kathmandu's monsoon climate. This points towards the need for features that explicitly capture seasonality and potentially more complex models to truly understand these interactions.
